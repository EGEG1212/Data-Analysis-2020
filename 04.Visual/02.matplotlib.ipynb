## 08. Matplotlib 그래프 영역 채우기
matplotlib.pyplot 모듈에서 그래프의 영역을 채우는 아래의 세가지 함수  
>fill_between()  
>fill_betweenx()  
>fill()  

```python
# fill_between() 사용(가로방향) 채우기
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_between(x[1:3], y[1:3], alpha=0.5)      ## fill_between() 사용(가로방향), alpha 투명값
# 네 점 (x[1], y[1]), (x[2], y[2]), (x[1], 0), (x[2], 0)을 잇는 영역이 채워짐...인덱스1~3번째 채워짐(2.4)(3.9)
plt.axhline(0); plt.axvline(0)  # 0라인이 그려짐 (없어도무방)
plt.show()
```

```python
# fill_betweenx() 사용(세로방향) 채우기
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_betweenx(y[2:4], x[2:4], color='pink', alpha=0.5)  ## fill_betweenx() 사용(세로방향)
# 네 점 (x[2], y[2]), (x[3], y[3]), (0, y[2]), (0, y[3])을 잇는 영역이 채워짐....인덱스2~4채워짐(3.9)(4.16)
plt.show()
```

```python
#두 그래프 사이 영역 채우기
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X-Label')  #선색상은 디폴트;
plt.ylabel('Y-Label')  #선색상은 디폴트;
plt.fill_between(x[1:3], y1[1:3], y2[1:3], color='lightgray', alpha=0.5)

plt.show()
```

```python
#두 그래프 사이 영역 채우기2_선을 곡선으로 만들기 (넘파이와 맷프롯리브 사용)
x = np.arange(1.0, 4.0, 0.01)  #1.0에서 4.0까지 간격0.02
y1 = x **2
y2 = x ** 2 / 2

plt.plot(x, y1); plt.plot(x, y2) #한줄로만들기
plt.xlabel('X-Label'); plt.ylabel('Y-Label')  #선색상은 디폴트;
plt.title('두 그래프 사이 영역채우기')
plt.fill_between(x[100:200], y1[100:200], y2[100:200], color='lightgray', alpha=0.5)

plt.show()
```

```python
# 임의의 영역 채우기
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 4, 8]

plt.plot(x, y1); plt.plot(x, y2)
plt.xlabel('X-Label'); plt.ylabel('Y-Label')
plt.title('임의의 영역채우기')
plt.fill([1.9, 1.9, 3.1, 3.1], [2, 5, 11, 8], color='lightgray', alpha=0.5)
# 4군데의 좌표를 찍어주면 됨............?

plt.show()
```

```python
# 임의의 영역 채우기2_5군데 찍어도 무방. 닫히기만 하면 됨 폴리곤?~
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 4, 8]

plt.plot(x, y1); plt.plot(x, y2)
plt.xlabel('X-Label'); plt.ylabel('Y-Label')
plt.title('임의의 영역채우기')
plt.fill([1.9, 1.9, 2.5, 3.1, 3.1], [2, 5, 9, 11, 8], color='lightgray', alpha=0.5)
# 5군데 찍어도 무방. 닫히기만 하면 됨 폴리곤?~ 임의의 모양 가능

plt.show()
```

## 10. Matplotlib 그리드 설정하기 plt.grid(True)
axis=y로 설정하면 가로 방향의 그리드만 표시

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True)

plt.show()
```

```python
# 축을 지정해서 할 수 있다. 축지정하기 
# {‘both’, ‘x’, ‘y’} 중 선택할 수 있고 디폴트는 ‘both’입니다.
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True, axis='y')
plt.title('축 지정하기(가로 y로만)')
plt.show()
```

```python
# 그리드의 스타일 설정
#color, alpha, linestyle 파마리터를 사용해서 그리드 선의 스타일을 설정했습니다.
#또한 which 파라미터를 ‘major’, ‘minor’, ‘both’ 등으로 사용하면 주눈금, 보조눈금에 각각 그리드를 표시할 수 있습니다.

a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True, axis='y', color='red', alpha=0.5, linestyle='--') #요부분
plt.title('그리드의 스타일설정')
plt.show()
```

## 11. Matplotlib 눈금 표시하기
틱 (Tick)은 그래프의 축에 간격을 구분하기 위해 표시하는 눈금  
matplotlib.pyplot 모듈의 xticks(), yticks(), tick_params() 함수를 이용

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.xticks([0, 1, 2])  #xticks(), yticks() 함수는 각각 x축, y축에 눈금을 표시. 
plt.yticks(np.arange(1, 6))

plt.show()
```

```python
#눈금 레이이블 지정하기
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])    #x축 0부터 2까지 0.2간격 눈금레이블을 명시적으로 지정
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))
     #y축 0부터 7까지 

plt.show()
```

```python
# 눈금 스타일 설정하기 .....이렇게까지는........ㅋ
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.tick_params(axis='x', direction='in', length=3, pad=6, labelsize=14, labelcolor='green', top=True)
plt.tick_params(axis='y', direction='inout', length=10, pad=15, labelsize=12, width=2, color='r')

plt.show()
```

tick_params() 함수를 사용하면 눈금의 스타일을 다양하게 설정할 수 있습니다.

axis는 설정이 적용될 축을 지정합니다. {‘x’, ‘y’, ‘both’} 중 선택할 수 있습니다.

direction을 ‘in’, ‘out’으로 설정하면 눈금이 안/밖으로 표시됩니다. {‘in’, ‘out’, ‘inout’} 중 선택할 수 있습니다.

length는 눈금의 길이를 지정합니다.

pad는 눈금과 레이블과의 거리를 지정합니다.

labelsize는 레이블의 크기를 지정합니다.

labelcolor는 레이블의 색상을 지정합니다.

top/bottom/left/right를 True/False로 지정하면 눈금이 표시될 위치를 선택할 수 있습니다.

width는 눈금의 너비를 지정합니다.

color는 눈금의 색상을 지정합니다.

## 12. Matplotlib 타이틀 설정하기
matplotlib.pyplot 모듈의 title() 함수를 이용

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True, axis='y', color='gray', alpha=0.5, linestyle='--')
plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=14)
plt.title('Sample graph', loc='right', pad=20) # 타이틀 위치와 오프셋 지정하기 (오른쪽,패딩20띄워서)

plt.show()
```

loc=’right’로 설정하면, 타이틀이 그래프의 오른쪽 위에 나타나게 됩니다.

‘left’, ‘center’, ‘right’로 설정할 수 있으며 디폴트는 ‘center’입니다.

pad=20은 타이틀과 그래프와의 간격 (오프셋)을 포인트 단위로 설정

```python
# 폰트 지정하기
title_font = {
    'fontsize': 16,
    'fontweight': 'bold',
    'color': 'magenta'
}
plt.title('Sample graph', fontdict=title_font, loc='left', pad=20)
```

```python
plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True, axis='y', color='gray', alpha=0.5, linestyle='--')
plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=14)
plt.title('Sample graph', loc='right', pad=20) 
plt.title('Sample graph', fontdict=title_font, loc='left', pad=20)

plt.show() # 네??????????????????
```

fontdict에 딕셔너리 형태로 폰트에 대한 설정을 입력할 수 있습니다.

‘fontsize’를 16으로, ‘fontweight’를 ‘bold’로 설정했습니다.

‘fontsize’는 포인트 단위의 숫자를 입력하거나 ‘smaller’, ‘x-large’ 등의 상대적인 설정을 할 수 있습니다.

‘fontweight’에는 ‘normal’, ‘bold’, ‘heavy’, ‘light’, ‘ultrabold’, ‘ultralight’의 설정을 할 수 있습니다.

## 13. Matplotlib 수직선/수평선 표시하기
matplotlib.pyplot 모듈에서는 아래의 네가지 함수를 지원합니다.  

axhline()  
axvline()  
hlines()  
vlines()  

```python
# 기본 사용 - axhline/axvline
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

#axhline() 함수의 첫번째 인자는 y 값으로서 수평선의 위치/두, 세번째 인자는 xmin, xmax 값으로서 0에서 1 사이의 값을 입력/0은 왼쪽 끝 (y축), 1은 오른쪽 끝을 의미
plt.axhline(1, 0, 0.55, color='gray', linestyle='--', linewidth='1')
plt.axvline(1, 0, 0.16, color='lightgray', linestyle=':', linewidth='2')
#axvline() 함수의 첫번째 인자는 x 값으로서 수직선의 위치/두, 세번째 인자는 ymin, ymax 값으로서 0에서 1 사이의 값을 입력/0은 아래쪽 끝 (x축), 1은 위쪽 끝을 의미


plt.axhline(5.83, 0, 0.95, color='gray', linestyle='--', linewidth='1')
plt.axvline(1.8, 0, 0.95, color='lightgray', linestyle=':', linewidth='2')

plt.show()
```

```python
# 기본 사용 - hlines/vlines
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

#hlines() 함수에 y, xmin, xmax를 순서대로 입력하면, 점 (xmin, y)에서 점 (xmax, y)를 따라 수평선을 표시합니다.
plt.hlines(4, 1, 1.6, colors='pink', linewidth=3)
plt.vlines(1, 1, 4, colors='pink', linewidth=3)
#vlines() 함수에 x, ymin, ymax를 순서대로 입력하면, 점 (x, ymin)에서 점 (x, ymax)를 따라 수평선을 표시

plt.show()
```

## 교수님문제 사인코사인을 그래프로 그리기

## 14. Matplotlib 막대 그래프 그리기   bar()
막대 그래프 (Bar graph, Bar chart)는 범주가 있는 데이터 값  
연속xx  

Matplotlib에서는 matplotlib.pyplot의 bar() 함수를 이용

```python
x = np.arange(3)  #주어진 범위와 간격에 따라 균일한 값을 갖는 어레이를 생성
years = ['2017', '2018', '2019']      #x축에 표시될 연도
values = [100, 400, 900]  #높이로 표시될 y 값

plt.bar(x, values) # plt.plot아니고 plt.bar이다  bar() 함수에 NumPy 어레이 [0, 1, 2]와 y 값에 해당하는 리스트 [100, 400, 900]를 입력
plt.xticks(x, years) #와 years를 입력해주면, x축에 ‘2017’, ‘2018’, ‘2019’가 순서대로 표시
plt.show()
```

```python
# 스타일꾸미기 (막대와 테두리의 색, 두께 등 다양한 스타일을 적용)
x = np.arange(3)
years = ['2017', '2018', '2019']
values = [100, 400, 900]

plt.bar(x, values, width=0.6, align='edge', color="springgreen",
        edgecolor="gray", linewidth=3, tick_label=years, log=True)
plt.show()
```

width는 막대의 너비입니다. 디폴트 값은 0.8이며, 예제에서는 0.6으로 설정했습니다.

align은 틱 (tick)과 막대의 위치를 조절합니다. 디폴트 값은 ‘center’인데, ‘edge’로 설정하면 막대의 왼쪽 끝에 x_tick이 표시됩니다.

color는 막대의 색을 지정합니다.

edgecolor는 막대의 테두리 색을 지정합니다.

linewidth는 테두리의 두께를 지정합니다.

tick_label을 어레이 형태로 지정하면, 틱에 어레이의 문자열을 순서대로 나타낼 수 있습니다.

log=True로 설정하면, y축이 로그 스케일로 표시

## 20. Matplotlib 파이 차트 그리기
파이 차트 (Pie chart, 원 그래프)는 범주별 구성 비율을 원형으로 표현한 그래프입니다.

위의 그림과 같이 부채꼴의 중심각을 구성 비율에 비례하도록 표현합니다.

matplotlib.pyplot 모듈의 pie() 함수를 이용

```python
ratio = [34, 32, 16, 18]  #모두 더하면 100이됨 
labels = ['Apple', 'Banana', 'Melon', 'Grapes'] # 3시부터 반시계방향으로 돌아감

plt.pie(ratio, labels=labels, autopct='%.1f%%')  #34.0%를 이렇게 표현
plt.show()
```

각 영역의 비율과 이름을 ratio와 labels로 지정해주고,

pie() 함수에 순서대로 입력합니다.

autopct는 부채꼴 안에 표시될 숫자의 형식을 지정합니다. 소수점 한자리까지 표시하도록 설정

```python
# 시작 각도와 방향 설정하기
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False)
# 260도부터 시작해서 counterclock=False시계방향으로 돌아감
plt.show()
```

startangle는 부채꼴이 그려지는 시작 각도를 설정합니다.

디폴트는 0도 (양의 방향 x축)로 설정되어 있습니다.

counterclock=False로 설정하면 시계 방향 순서로 부채꼴 영역이 표시

```python
# 중심에서 벗어나는 정도 설정하기*****************많이쓰일듯한 
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.10, 0, 0.30] # 0.1만큼 떨어지기 explode터뜨리다 boom!!!!!!!

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode)
plt.show()
```

explode는 부채꼴이 파이 차트의 중심에서 벗어나는 정도를 설정합니다.

‘Banana’와 ‘Grapes’ 영역에 대해서 반지름의 10% 만큼 벗어나도록 설정

```python
#교수님ppt예제 _여러 개의 그래프를 그리는 방법(Subplot) 1
def f(t):
 return np.exp(-t) * np.cos(2*np.pi*t)
def g(t):
 return np.sin(np.pi*t)
t1 = np.arange(0.0, 5.0, 0.01)
t2 = np.arange(0.0, 5.0, 0.01)
plt.subplot(221) # 2 x 2, 그래프중 첫번째
plt.plot(t1, f(t1))
plt.subplot(2, 2, 2)
plt.plot(t2, g(t2))
plt.subplot(223)
plt.plot(t1, f(t1), 'r-')
plt.subplot(224)
plt.plot(t2, g(t2), 'r-')
plt.show()
```

```python
#교수님ppt예제 _여러 개의 그래프를 그리는 방법(Subplot) 2 > 03.고급편에서 풀이함
import seaborn as sns
ans = sns.load_dataset('anscombe')
ans.shape; ans.head(); ans.tail()
ans.groupby('dataset').describe()
ds1 = ans[ans['dataset'] == 'I']; ds2 = ans[ans['dataset'] == 'II']
ds3 = ans[ans['dataset'] == 'III']; ds4 = ans[ans['dataset'] == 'IV']
fig = plt.figure()
axes1 = fig.add_subplot(2, 2, 1); axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3); axes4 = fig.add_subplot(2, 2, 4)
axes1.plot(ds1['x'], ds1['y'], 'o'); axes2.plot(ds2['x'], ds2['y'], 'o')
axes3.plot(ds3['x'], ds3['y'], 'o'); axes4.plot(ds4['x'], ds4['y'], 'o')
axes1.set_title('Dataset 1'); axes2.set_title('Dataset 2')
axes3.set_title('Dataset 3'); axes4.set_title('Dataset 4')
fig.suptitle("Anscombe's Quartet")
fig.tight_layout()
```
