#### 1-b. 3분위수(Q3)와 1분위수(Q1)의 차이보다 1.5배가 크거나 작은 데이터는 이상치이다.  
즉, Q1 - 1.5 * (Q3 - Q1) 보다 작은 데이터  
   Q3 + 1.5 * (Q3 - Q1) 보다 큰 데이터  
   이 이상치를 제거하고 위의 4가지 항목에 대해서 평균, 표준편차를 구하시오.

```python
def get_outlier(Series):
    q1 = np.percentile(Series, 25)
    q3 = np.percentile(Series, 75)
    
    IQR_weight = (q3-q1)*1.5
    low_bound, high_bound = q1 - IQR_weight, q3 + IQR_weight
    
    outlier_idx = np.where((Series < low_bound) | (Series > high_bound))
    return outlier_idx
```

## 2. Titanic - sns.load_dataset('titanic')  
#### 2-a. 타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category1열을 만드시오.  
category1 다음과 같이 정의  
1) 20살이 넘으면 성별을 그대로 사용  
2) 20살 미만이면 성별에 관계없이 "child"라고 한다.  

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic["category1"] = titanic.apply(lambda r: r.sex if r.age >= 20 else "child", axis=1)
titanic.tail()
```

#### 2-b. 타이타닉호의 승객 중 나이를 명시하지 않은 고객은 나이를 명시한 고객의 평균 나이 값이 되도록 titanic 데이터프레임을 고치시오.

```python
titanic['age'] = titanic['age'].fillna(titanic['age'].mean())
titanic.tail()
```

#### 2-c. 성별, 선실(class)별, 출발지(embark_town)별 생존율
(alive범주형 survived:정수형)

```python
titanic["survived"].value_counts()
```

```python
titanic.pivot_table('survived',['sex', 'class', 'embark_town'])  
```

```python
# 선실 대 성별
print(titanic.pivot_table(index='sex', columns='pclass', values='survived', aggfunc=np.mean), "\n")
```

#### 2-d. 타이타닉호 승객을 '미성년자', '청년', '중년', '장년', '노년' 나이 그룹으로 나누고, 각 그룹별 생존율  

```python
bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "중년", "장년", "노년"] 
titanic["age_cat"] = pd.cut(titanic.age, bins, labels=labels)
titanic.tail()
```

```python
titanic.pivot_table('survived', 'age_cat') 
```

#### 2-e. qcut명령으로 세 개의 나이 그룹을 만들고, 나이 그룹별 남녀 성비와 생존율

```python
titanic['age_group'] = pd.qcut(titanic.age, 3, labels=['A1', 'A2', 'A3'])
titanic.tail()
```

```python
titanic.pivot_table('survived', 'age_group')
```

```python
# gender: 남성이면 1, 여성이면 0
titanic['gender'] = titanic.apply(lambda r: 1 if r.sex == 'male' else 0, axis=1)
titanic.tail()
```

```python
#같은결과1 pivot_table
titanic.pivot_table(['survived','gender'], 'age_group')  
```

```python
#같은결과2 groupby
titanic[['age_group', 'gender', 'survived']].groupby(['age_group']).mean()
```

## 3. Mile Per Gallon - sns.load_dataset('mpg')

```python
mpg = sns.load_dataset('mpg')
mpg.tail()
```

```python
mpg.describe()  #398개정도군..
```

#### 3-a. 배기량(displacement) 대비 마력(horsepower) 열(hp_per_cc)을 추가 

```python
mpg["hp_per_cc"] = mpg.horsepower / mpg.displacement
mpg.head()
```

#### 3-b. name으로부터 manufacturer(제조사)와 모델을 추출.  새로운 열 manufacturer와 model을 추가. name열은 삭제 

```python
def get_mfger(full_name):
    return full_name.split()[0] #스플릿에 기본빈칸스페이스. 리스트의 첫번째[0]을 부르자:chrysler
def get_model(full_name):
    return ' '.join(full_name.split()[1:]) #이후 1부터 끝까지 스페이스로 조인함.모델명
```

```python
get_mfger('chrysler lebaron medallion'), get_model('chrysler lebaron medallion')
```

```python
get_mfger('ford torino'), get_model('ford torino')
```

```python
# @_@
mpg['manufacturer'] = mpg.name.apply(get_mfger)
mpg['model'] = mpg.name.apply(get_model)
del mpg['name']
mpg.head()
```

#### 3-c. 에진의 실린더(cylinders) 갯수별 연비(mpg)의 평균  

```python
#같은결과 groupby
mpg[['cylinders','mpg']].groupby('cylinders').mean()
```

```python
#같은결과 pivot_table
mpg.pivot_table('mpg','cylinders')
```

#### 3-d. 생산지(origin)별 배기량 대비 마력(hp_per_cc)의 평균  

```python
mpg[['origin', 'hp_per_cc']].groupby('origin').mean()
```

#### 3-e. 모델이 5개 이상인 제조사에 대하여 연비(mpg)의 평균이 가장 좋은 제조사 Top5를 구하시오. 

```python
mpg.manufacturer.value_counts()
```

```python
s = mpg.manufacturer.value_counts() #위와같은 목롥을 s에 넣어
big = s[s >= 5].index.tolist()   # 5인 이상인 것만 tolist()
print(big)
```

```python
# in 연산자 대신에 isin() method 사용
mpg[mpg['manufacturer'].isin(big)].head()
```

```python
#같은결과 groupby
mpg[mpg['manufacturer'].isin(big)][['manufacturer','mpg']].groupby('manufacturer').mean()
```

```python
#같은결과 pivot_table
mpg[mpg['manufacturer'].isin(big)].pivot_table('mpg','manufacturer')
```

```python
#정렬,top5[:5]
mpg[mpg['manufacturer'].isin(big)].pivot_table('mpg','manufacturer').sort_values(by='mpg', ascending=False)[:5]
```
