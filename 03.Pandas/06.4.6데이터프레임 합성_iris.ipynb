## 4.6 데이터프레임 합성

```python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df1
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2
```

```python
# 위의 두 데이터는.. 고객번호
#좌 1002 1003 1004 1007
#교집합 1001 1005 1006
#우 1008
```

```python
# inner join 
pd.merge(df1, df2)  
#컬럼명이 같은 기준으로 알아서 머지해줌??우와 똑똑쓰. = 인너조인과같음
#교집합 1001 1005 1006에 대한 데이터가 나왔다.
```

```python
# outer join  키 값이 한쪽에만 있어도 데이터를 보여준다.
pd.merge(df1, df2, how='outer')
```

```python
#left outer join
pd.merge(df1, df2, how='left')  #left df1에 대한 부분 출력 (1008 제외)
```

```python
#right outer join
pd.merge(df1, df2, how='right') #right df2에 대한 부분 출력 (1002.1003.1004.1007 제외)
```

```python
# 이건 좀 헷갈리는 부분 행부분이 갯수가 다른경우..
df1 = pd.DataFrame({
    '품종': ['setosa', 'setosa', 'virginica', 'virginica'],
    '꽃잎길이': [1.4, 1.3, 1.5, 1.3]},
    columns=['품종', '꽃잎길이'])
df1
```

```python
df2 = pd.DataFrame({
    '품종': ['setosa', 'virginica', 'virginica', 'versicolor'],
    '꽃잎너비': [0.4, 0.3, 0.5, 0.3]},
    columns=['품종', '꽃잎너비'])
df2
```

```python
pd.merge(df1, df2)   #setosa 두가지조합, virginica 네가지조합 독특...
```

```python
## 컬럼명이 다른 경우  (고객명만 같으면 문제없는데) 고객명도, 데이터도 같아서..... 
df1 = pd.DataFrame({
    '고객명': ['춘향', '춘향', '몽룡'],
    '날짜': ['2018-01-01', '2018-01-02', '2018-01-01'],
    '데이터': ['20000', '30000', '100000']})
df1
```

```python
df2 = pd.DataFrame({
    '고객명': ['춘향', '몽룡'],
    '데이터': ['여자', '남자']})
df2
```

```python
pd.merge(df1, df2)  # 고객명도, 데이터도 칼럼이 같으니까..  결과값이 안나옴.
```

```python
pd.merge(df1, df2, on='고객명')  #SQL처럼 고객명기준으로 묶어라! 하니까.  알아서데이터를 두개로 구분지어줌
```

```python
df1 = pd.DataFrame({
    '이름': ['영희', '철수', '철수'],
    '성적': [1, 2, 3]})
df1
```

```python
df2 = pd.DataFrame({
    '성명': ['영희', '영희', '철수'],
    '성적2': [4, 5, 6]})
df2
```

```python
pd.merge(df1, df2, left_on='이름', right_on="성명")
```

```python
# 지금까지 컬럼으로 머지했다면
# 아래부터는 인덱스로 머지하겠다!! [인덱스를 기준열로 사용]
```

```python
df1 = pd.DataFrame({
    '도시': ['서울', '서울', '서울', '부산', '부산'],
    '연도': [2000, 2005, 2010, 2000, 2005],
    '인구': [9853972, 9762546, 9631482, 3655437, 3512547]})
df1
```

```python
df2 = pd.DataFrame(
    np.arange(12).reshape((6, 2)),
    index=[['부산', '부산', '서울', '서울', '서울', '서울'],
           [2000, 2005, 2000, 2005, 2010, 2015]],
    columns=['데이터1', '데이터2'])
df2  #부산서울, 년도별 인덱스가 있는 다중인덱스
#서울4건
```

```python
pd.merge(df1, df2, left_on=['도시', '연도'], right_index=True)
#왼쪽은 도시와 연도를.. 
#오른쪽은 인덱스를 트루로 놓아서 merge함.
```

## join메서드 (인덱스로 합쳐지나 봄)

```python
df1.join(df2, how='outer')  #에러.. ㅋㅋㅋ 
```

## concat 함수를 사용한 데이터 연결
기준 열(key column)을 사용하지 않고 단순히 데이터를 연결(concatenate)한다.

```python
s1 = pd.Series([0, 1], index=['A', 'B'])
s2 = pd.Series([2, 3, 4], index=['A', 'B', 'C'])
```

```python
s1
```

```python
s2
```

```python
pd.concat([s1, s2])  # 시리즈로 합쳐질뿐.. 위아래로 딸려 들어감? vstack같은건가???
```

```python
# 옆으로 데이터 열을 연결하고 싶으면 axis=1로 인수를 설정
df1 = pd.DataFrame(
    np.arange(6).reshape(3, 2),
    index=['a', 'b', 'c'],
    columns=['데이터1', '데이터2'])
df1
```

```python
df2 = pd.DataFrame(
    5 + np.arange(4).reshape(2, 2),
    index=['a', 'c'],
    columns=['데이터3', '데이터4'])
df2
```

```python
pd.concat([df1, df2], axis=1)  
#옆으로(행방향) 데이터 열을 연결하고 싶으면 axis=1로 인수를 설정
#리스트로 주기[df1, df2]
```

```python
pd.concat([df1, df2], axis=0)   # 위와 이거랑 다르지영? 행으로 연결
```
