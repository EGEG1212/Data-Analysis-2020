## cctv 데이터 파악하고 정리하기

```python
cctv = pd.read_excel('./data/서울시 자치구 년도별 CCTV 설치 현황(2011년 이전_2018년).xlsx')
cctv.head()
```

```python
# Nan 이 있는지 확인
cctv.isnull().values.any  
# nan 있다 True (동대문구, 중랑구)
```

```python
# Nan에 0을 넣어라
cctv.fillna(0, inplace=True)
cctv.head()
```

```python
# Nan 이 있는지 확인하는 또다른 방법
cctv.isnull().any().any()
```

```python
# 컬럼 '2015년_이전' 추가 (값은 2011+2012+2013+2014+2015 더해서 넣음)
cctv.rename({'기관명':'구별'}, inplace=True, axis=1) # 컬럼명 '기관명'을 '구별'로 변경
cctv['2015년_이전'] = cctv['2011년 이전'] + cctv['2012년'] + cctv['2013년'] + cctv['2014년'] + cctv['2015년']
cctv.head()
```

```python
# 구별의 '강 남 구'의 빈칸을 없애기.map(lambda s: s.replace(' ', ''))
cctv['구별'] = cctv['구별'].map(lambda s: s.replace(' ', ''))
cctv.head()
```

```python
#컬럼 구별의 내용 유니크한지, 중복되는건 없는지  확인
cctv['구별'].unique()
```

```python
del cctv['2011년 이전']
del cctv['2012년']
del cctv['2013년']
del cctv['2014년']
del cctv['2015년']
cctv = cctv [['구별', '소계', '2015년_이전', '2016년', '2017년', '2018년']]
cctv.tail()
```

```python
# 컬럼 2015년_이전 부분이 1841.0이었는데 1841로 바뀌었다! (int로변경)
cctv = cctv.astype({'2015년_이전': int})
cctv.tail()
```

```python
# 최근 CCTV설치 3년간 증가율
cctv['최근증가율'] = (cctv['2016년'] + cctv['2017년'] + cctv['2018년']) / cctv['2015년_이전'] * 100
cctv.tail()
```

```python
# 최근 3년간 CCTV설치 증가율이 높은 구_[TOP5] 중구, 영등포구, 성동구, 금천구, 광진구
cctv.sort_values('최근증가율', ascending=False).head() #내림차순ascending=False
```

```python
# cctv대수가 작은 구
cctv.sort_values('소계').head()   #오름차순ascending=True 디폴트
#[battom5] 도봉구, 중랑구, 강북구, 종로구, 중구
#도봉구가 제일적다, 안전한동네인가? cctv설치할 자본이 없는건가?
```

```python
# cctv대수가 많은 구
cctv.sort_values('소계', ascending=False).head()
#[TOP5]강남구, 관악구, 구로구, 성북구, 은평구
```

## 서울시 인구 현황 데이터 파악하고 정리하기

```python
pop = pd.read_csv('./data/서울시 인구현황(2020 3분기).tsv', sep='\t', header=2)
pop.head()
```

```python
#김태우씨코드 (txt 파일을 csv로 읽는데 delimiter를 '\t'로 두면 됨)
#필요한것만 쏙쏙골라 열었다
pop_Seoul = pd.read_csv('./data/서울시 인구현황(2020 3분기).txt', delimiter = '\t',header=2)[['자치구','계','계.1','계.2','65세이상고령자']]
pop_Seoul.head()
```

```python
del pop['기간']
del pop['세대']
del pop['남자']
del pop['남자.1']
del pop['남자.2']
del pop['여자']
del pop['여자.1']
del pop['여자.2']
del pop['세대당인구']
pop.head()
```

```python
# 컬럼명 변경
pop.rename({'자치구': '구별', '계':'인구수', '계.1':'내국인', '계.2':'외국인', '65세이상고령자':'고령자'}, inplace=True, axis=1)
pop.head()
```

```python
# ,없애고 int적용
pop['인구수'] = pop['인구수'].map(lambda s: int(s.replace(',','')))
pop['내국인'] = pop['내국인'].map(lambda s: int(s.replace(',','')))
pop['외국인'] = pop['외국인'].map(lambda s: int(s.replace(',','')))
pop['고령자'] = pop['고령자'].map(lambda s: int(s.replace(',','')))
pop.head()
```

```python
# 합계행 제거
pop.drop([0], inplace=True) 
pop.head()
```

```python
pop['구별'].unique()
```

```python
# 컬럼추가 외국인비율, 고령자비율
pop['외국인비율'] = pop['외국인'] / pop['인구수'] * 100
pop['고령자비율'] = pop['고령자'] / pop['인구수'] * 100
pop.tail()
```

```python
# 외국인 비율이 높은 구 [TOP5]역시 영등포, 중구, 금천구, 구로구, 용산구
pop.sort_values('외국인비율', ascending=False).head()
```

```python
# 고령자 비율이 높은 구 [TOP5]강북구, 도봉구, 중구, 종로구, 은평구
pop.sort_values('고령자비율', ascending=False).head()
```

## CCTV 데이터와 인구 현황 데이터를 합치고 분석하기

```python
df = pd.merge(cctv, pop, on='구별')
df.head()
```

```python
del df['2015년_이전']
del df['2016년']
del df['2017년']
del df['2018년']
df.tail()
```

```python
# '구별'부분을 인덱스로 삼자
df.set_index('구별', inplace=True)
df.head()
```

### 상관관계분석 np.corrcoef
얼마나 상관이 있는지 알아보는거..?  
점선(대각선)의 기울기가..  
1에 가까울수록 y축에 붙고  
0에 가까울수록 x축에 붙는다.  
0.5에 가까울수록 상관이 많이 된다?   

```python
#고령자 비율과 CCTV댓수
np.corrcoef(df['고령자비율'], df['소계'])  #음의상관관계
```

```python
# 외국인 비율과 CCTV댓수
np.corrcoef(df['외국인비율'], df['소계'])
```

```python
#인구수와 cctv 댓수
np.corrcoef(df['인구수'], df['소계']) #양의상관관계
```

## CCTV 현황 그래프로 분석하기

```python
df['소계'].plot(kind='barh', grid=True, figsize=(12,8), title='구별 CCTV 댓수')
plt.show()
```

```python
# 값으로 정렬하기  sort_values()
df['소계'].sort_values().plot(kind='barh', grid=True, figsize=(12,8), title='구별 CCTV 댓수')
plt.show()
```

```python
# 인구대비 CCTV비율
# 계산해서 바로 바그래프 그림
df['cctv비율'] = df['소계'] / df['인구수'] * 100
df['cctv비율'].sort_values().plot(kind='barh', grid=True, figsize=(12,8))
plt.title('인구대비 CCTV 비율')
plt.show()
```

```python
# 인구수와 CCTV 댓수의 산점도
plt.figure(figsize=(6,6))
plt.scatter(df['인구수'], df['소계'], s=50)
plt.grid(True)
plt.title('인구수와 CCTV 댓수의 산점도')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.show()
```

```python
# 산점도 위에 상관관계를 나타내는 선 그리기 ??????????응???? polyfit 이게 뭐지? np.corrcoef 비슷하게 생겼다아  ㅋ.ㅋ
# polyfit 직선그리기 
fp1 = np.polyfit(df['인구수'], df['소계'], 1)
fp1
```

```python
fx = np.array([100000, 700000])  #x축은 numpy의 linspace로 만들고
f1 = np.poly1d(fp1)              #y축은 poly1d로 만듦
fy = f1(fx)
fy
```

```python
plt.figure(figsize=(8,8))
plt.scatter(df['인구수'], df['소계'], s=50)
plt.plot(fx, fy, ls='dashed', lw=3, color='g') # 그린점선:전체 데이터의 대표값 역할을 한다. 
plt.grid(True)
plt.title('인구수와 CCTV댓수의 관계')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.show()
```

```python
# 직선이 이 전체 데이터의 대표값 역할을 한다. 
# 직선에서 멀어질수록 다른 색을 나타내고싶어서
# 오차를 계산할 수 있는 코드를 만들고 오차큰순으로 데이터 정렬
# np.abs ??????
df['오차'] = np.abs(df['소계'] - f1(df['인구수']))
df_sort = df.sort_values('오차', ascending=False)
df_sort.head()
```

```python
# 텍스트와 color map 입히기
plt.figure(figsize=(14,10))
plt.scatter(df['인구수'], df['소계'], c=df['오차'], s=50)
plt.plot(fx, fy, ls='dashed', lw=3, color='g')

for i in range(10):
    plt.text(df_sort['인구수'][i]*1.02, 
                df_sort['소계'][i]*0.98,
                df_sort.index[i], fontsize=15)
plt.grid(True)
plt.title('인구수와 CCTV 댓수의 관계', fontsize=20)
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.colorbar()
plt.show()

#강남구, 관악구, 구로구, 성동구는 서울시 전체의 지역의 일반적인 경향보다 CCTV가 많이 설치 된 지역
#송파구, 강서구, 노원구, 중랑구, 도봉구, 강북구는 일반적인 경향보다 CCTV가 적게 설치된 지역
```
