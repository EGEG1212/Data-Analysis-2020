```python
import numpy as np
import pandas as pd
import folium 
```

```python
df = pd.read_csv('gscafe1.csv')
df.tail()
```

```python
mapping = folium.Map(location=[df.위도.mean(), df.경도.mean()], zoom_start=13)
#강서구 주소지만있는데 왜 안성으로가는지... 부산강서구포함인듯 ㅋ
mapping
```

```python
for i in df.index:
    folium.Marker(
    location=[df.위도[i], df.경도[i]],
    popup=df.상호명[i],
    tooltip=f'{df.건물명[i]}',
    icon=folium.Icon(color='blue', icon='map-marker')
).add_to(mapping)
mapping  #한글이 깨져서 '상호명,건물명' 확인불가 ㅋㅋㅋ
```
