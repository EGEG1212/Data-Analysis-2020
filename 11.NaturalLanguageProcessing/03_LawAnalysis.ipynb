## 육아휴직 관련 법안 분석

```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib as mpl 
import matplotlib.pyplot as plt 
%matplotlib inline
# 한글폰트 사용
mpl.rc('font', family='Malgun Gothic')
mpl.rc('axes', unicode_minus=False)
```

```python
import nltk
from konlpy.tag import Okt
from konlpy.corpus import kobill
```

```python
files_ko = kobill.fileids()
doc_ko = kobill.open('1809890.txt').read()
```

```python
doc_ko[:1000]  #1000 글자만
```

```python
okt = Okt()
tokens_ko = okt.nouns(doc_ko)  #명사추출nouns
tokens_ko[:10]
```

```python
#(기존 k였고 이젠 NLTK)
#수집된 단어의 횟수 및 고유한 횟수
ko = nltk.Text(tokens_ko, name='대한민국 국회 의안 제 1809890호')
len(ko.tokens), len(set(ko.tokens)) #이렇게 적은 단어를 썼는지의심...735단어
```

```python
ko.vocab() # 단어빈도수
```

```python
plt.figure(figsize=(15,8))
ko.plot(50)
plt.show()
```

```python
stop_words = ['.', '(', ')', ',', "'", '%', '-', 'X', ').', '×','의','자','에','안','번',
                      '호','을','이','다','만','로','가','를']      #노가다로 많이 넣으면 넣을수록 품질이 좋아짐
ko = [each_word for each_word in ko if each_word not in stop_words] #필터링; stop_words가 아닌놈만 골라내겠다.
print(ko[:10])
```

```python
#stop_words를 제거한걸 가지고 그려보는 그래프
ko = nltk.Text(ko, name='대한민국 국회 의안 제 1809890호')
plt.figure(figsize=(15,6))
ko.plot(50)     # Plot sorted frequency of top 50 tokens
plt.show()  #끝에 단어가 조금 바뀌었다.
```

```python
#초등학교가 문서 내에서 몇 번 언급되었나
ko.count('초등학교')  #6번
```

```python
# 원하는 단어의 문서내 대략적인 위치와 분량 확인
plt.figure(figsize=(12,6))
ko.dispersion_plot(['육아휴직', '초등학교', '공무원'])
```

```python
#원하는 단어의 주변부 단어 확인
ko.concordance('초등학교')
```

```python
#연어
ko.collocations()
```

```python
from wordcloud import WordCloud
data = ko.vocab().most_common(150)   #150단어만 폰트패스지정

wordcloud = WordCloud(font_path='c:/Windows/Fonts/malgun.ttf',
                      relative_scaling = 0.2,
                      background_color='white',
                      width=1000, height=600
                      ).generate_from_frequencies(dict(data))
plt.figure(figsize=(12,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
