```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib as mpl 
import matplotlib.pyplot as plt 
%matplotlib inline
# 한글폰트 사용
mpl.rc('font', family='Malgun Gothic')
mpl.rc('axes', unicode_minus=False)
```

```python
# 빈 데이터프레임을 만들어주고
benz = pd.DataFrame(columns=['상호명', '지점명', '도로명주소', '경도', '위도'])
bmw = pd.DataFrame(columns=['상호명', '지점명', '도로명주소', '경도', '위도'])
volkswagen = pd.DataFrame(columns=['상호명', '지점명', '도로명주소', '경도', '위도'])
porsche = pd.DataFrame(columns=['상호명', '지점명', '도로명주소', '경도', '위도'])
```

```python
# 17개 파일 모두
from glob import glob
data_files = glob('./data/소상공인시장진흥공단_상가(상권)정보*.csv')

for filename in data_files:
    print(filename)
    df = pd.read_csv(filename, sep='|')  #파일읽고
    df = df[(df['상권업종중분류명'] == '자동차/자동차용품') | (df['상권업종중분류명'] == '개인/가정용품수리') | (df['상권업종중분류명'] == '자동차/이륜차')]
    df = df[(df['상권업종중분류코드'] == 'D23') | (df['상권업종중분류코드'] == 'F03') | (df['상권업종중분류코드'] == 'F14')]  #분류코드3개를 df에 넣고
    df = df[['상호명', '지점명', '도로명주소', '경도', '위도']] #데이터프레임에

    # 컬럼상호명에 지정한str이 대소문자구별없이(case=False) 포함되어있다면, 각각변수에 저장
    bz = df[df['상호명'].str.contains('벤츠|benz', case=False)]
    bm = df[df['상호명'].str.contains('비엠더블유|BMW', case=False)]
    vw = df[df['상호명'].str.contains('폭스바겐', case=False)]
    ps = df[df['상호명'].str.contains('포르쉐', case=False)]

    # 빈 데이터프레임과 각각 합치기concat
    benz = pd.concat([benz, bz])
    bmw = pd.concat([bmw, bm])
    volkswagen = pd.concat([volkswagen, vw])
    porsche = pd.concat([porsche, ps])
```

```python
# 결과저장
benz.to_csv('./results/벤츠.csv', index=False)
bmw.to_csv('./results/비엠더블유.csv', index=False)
volkswagen.to_csv('./results/폭스바겐.csv', index=False)
porsche.to_csv('./results/포르쉐.csv', index=False)
```

```python
#mapUtil.py모듈활용_도로명주소에서 광역시도와 시군고/행정구명 찾기_벤츠
benz.reset_index(inplace=True)     #리셋인덱스(인덱스추가됨)
benz.drop('index', axis=1, inplace=True)# 추가된인덱스(인덱스2줄) 중 한개 드롭.axis=1줘서 컬럼삭제
benz['ID'] = mu.get_ID(benz)
benz['브랜드명'] = 'benz'
benz['benz 매장수'] = 1
benz.head()
```

```python
#mapUtil.py모듈활용_도로명주소에서 광역시도와 시군구/행정구명 찾기_커피빈
bmw.reset_index(inplace=True)
bmw.drop('index', axis=1, inplace=True)
bmw['ID'] = mu.get_ID(bmw)
bmw['브랜드명'] = 'BMW'
bmw['BMW 매장수'] = 1
bmw.head()
```

```python
#mapUtil.py모듈활용_도로명주소에서 광역시도와 시군구/행정구명 찾기_폭스바겐
volkswagen.reset_index(inplace=True)
volkswagen.drop('index', axis=1, inplace=True)
volkswagen['ID'] = mu.get_ID(volkswagen)
volkswagen['브랜드명'] = 'VW'
volkswagen['VW 매장수'] = 1
volkswagen.head()
```

```python
#mapUtil.py모듈활용_도로명주소에서 광역시도와 시군구/행정구명 찾기_빽다방
porsche.reset_index(inplace=True)
porsche.drop('index', axis=1, inplace=True)
porsche['ID'] = mu.get_ID(porsche)
porsche['브랜드명'] = '포르쉐'
porsche['포르쉐 매장수'] = 1
porsche.head()
```

### 브랜드별 피벗테이블 생성

```python
pt_bz = benz.pivot_table('benz 매장수', index='ID', aggfunc='sum')
pt_bz = pt_bz.reset_index()
print(len(pt_bz))
pt_bz.head()
```

```python
pt_bm = bmw.pivot_table('BMW 매장수', index='ID', aggfunc='sum')
pt_bm = pt_bm.reset_index()
print(len(pt_bm))
pt_bm.head()
```

```python
pt_vw = volkswagen.pivot_table('VW 매장수', index='ID', aggfunc='sum')
pt_vw = pt_vw.reset_index()
print(len(pt_vw))
pt_vw.head()
```

```python
pt_ps = porsche.pivot_table('포르쉐 매장수', index='ID', aggfunc='sum')
pt_ps = pt_ps.reset_index()
print(len(pt_ps))
pt_ps.head()
```

## 카페 데이터 병합

```python
car = pd.merge(pt_bz, pt_bm, how='left', on='ID')
car.head()
```

```python
car = pd.merge(car, pt_vw, how='left', on='ID')
car = pd.merge(car, pt_ps, how='left', on='ID')
car.head()
```

```python
car = car.fillna(0)
car.head()
```

```python
car['BENZ 매장수'] = car['benz 매장수'].astype(int)
car['BMW 매장수'] = car['BMW 매장수'].astype(int)
car['VW 매장수'] = car['VW 매장수'].astype(int)
car['포르쉐 매장수'] = car['포르쉐 매장수'].astype(int)
```

```python
#컬럼 총 매장수 추가 
car['합계'] = (car['BENZ 매장수']+car['BMW 매장수'])+(car['VW 매장수']+car['포르쉐 매장수'])
car.tail()
```
